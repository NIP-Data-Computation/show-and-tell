# Investigating the capitalist tactic on the new 'supergroup' project of SM Entertainment
\
SuperM is a "supergroup" formed by SM Entertainment (which is the biggest company in the K-Pop industry), consisting of members from four groups under the company: Taemin from **SHINee**, Baekhyun and Kai from **EXO**, Taeyong and Mark from **NCT 127** and Ten and Lucas from **WayV** (just a Chinese sub-unit of NCT). The company aims to put together members that have established a "star" in their own bands. Thus, these members are the most popular ones in their own respective groups. The company planned to debut them as a "K-Pop Super Boy Band" and was described as the "Avengers of K-Pop" by *Billboard* and *Los Angeles Times*.

## Theories
*   Increase sales and popularity to lesser active groups.
*   Increase more brand reputation to the groups involved.
*   The project will backfire because of 'too many' popular members.
*   They got threatened by BTS' popularity and attempted the steps of BTS but failed.





### Methodology
1.   Data Gathering
  *   International Album Sales
  > The research head of the Korean Music Content Association (KMCA), which is in-charge of the Gaon Chart, was contacted, along with the biggest international K-Pop Merchandise company, Ktown4U. Unfortunately, there were no replies from the KMCA, but luckily, the constituents of Ktown4U offered help. The international sales are still valid since SuperM debuted in the US/Western market.
  > ![](https://drive.google.com/uc?export=view&id=1-KM-3j7eRAMXLm6dad967rPnfSvaoAc5)
  *   Brand Reputation
  > The K-Pop Brand Reputation Rankings are released by the Korean Business Research Institute (KBRI). They usuallly just release the Top 30 of each category which they deem the currently relevant idols or groups per month. The Brand Reputation consists of four parts.
  > ![](https://drive.google.com/uc?export=view&id=1yxZS8GDOX1V6BOXU4Ia1230jpivxRPSQ)
      *   Participation Index
        >> How much the idol or a group was searched on different sites (e.g. Wikipedia, Navor, Daum, Google, and etc.)
      *   Media Index
        >> How much people shared, engaged, or created articles about the idols or groups on different sites (e.g. Yahoo News, CNN Entertainment, The Korea Times, The Chosun Ilbo, etc.).
      *   Communication/Interaction Index
        >> How much people created blogs about the idols/groups and engaged with it on different sites/platforms (e.g. commenting, giving likes to the blog posts, example is a Korean blogging site called 'pannchoa', Tumblr, Reddit, etc.)
      *   Community Index
        >> How much people mentioned or engaged with them on social media platforms (e.g. Twitter, Facebook, Instagram, TikTok, etc.).

>**Theory no. 4: They got threatened by BTS' popularity and attempted the steps of BTS but failed.**
\
All of the international album sales was taken into account for this section. The old groups were added into the ```fourth_old``` DataFrame and relatively newer groups were added to the ```fourth_new``` DataFrame.

````
fourth_old = pd.read_excel(filepath, sheet_name = "SHINee", usecols = ["Month", "Album Sales"])


for name in group_name:
  fourth_old[name] = pd.read_excel(filepath, sheet_name = name, usecols = ["Album Sales"])

print(fourth_old.head())
print(fourth_old.dtypes)
````

````
#Renaming the column name to the group name.
fourth_old = fourth_old.rename(columns={"Album Sales" : "SHINee"})
print(fourth_old.head())

#Parsing the Month column.
fourth_old['Month'] =  pd.to_datetime(fourth_old['Month'], format='%Y-%m')
print(fourth_old.dtypes)
````

>> Creating the ```fourth_new``` DataFrame.

````
fourth_new = pd.read_excel(filepath, sheet_name="SuperM", usecols=["Month", "Album Sales"])
fourth_new["WAYV"] = pd.read_excel(filepath, sheet_name="WAYV", usecols=["Album Sales"])

#Renaming the column name to the group name.
fourth_new = fourth_new.rename(columns={"Album Sales" : "SuperM"})
print(fourth_new.head())

#Parsing the Month column.
fourth_new['Month'] =  pd.to_datetime(fourth_new['Month'], format='%Y-%m')
print(fourth_new.dtypes)
````

> **Performing a Left Join to merge the New groups to the Old groups DataFrame**

````
fourth = pd.merge(fourth_old, fourth_new, on=["Month"], how = "left")
print(fourth)
````

>**Normalizing the Data**
\
The album sales were normalized because of the huge gap in the data. Moreover, normalizing the newer groups as is returns all values as NaN thus, ```.iloc``` was used instead.

````
def norm(data):
  return (data)/(max(data)-min(data))

for name in ["SHINee", "EXO", "NCT"]:
  fourth[name] = norm(fourth[name])

fourth.iloc[9:20, 4] = norm(fourth.iloc[9:20, 4])
fourth.iloc[9:20, 5] = norm(fourth.iloc[9:20, 5])

print(fourth)
````

>**Plotting the normalized data.**
\
The progressive increase of album sales of SuperM can be observed when bigger and more popular groups such as NCT and EXO made their individual activities. This might imply that SuperM might lifted up popularity of groups on hiatus, it still dependent on the popularity of the active predecessor groups where the members came from.

````
plt.clf()

fig,ax = plt.subplots()

def plot_subplots(axes, x, y, color, style, label):
  axes.plot(x, y, color = color, linestyle = style, label = label)

plot_subplots(ax, fourth["Month"], fourth["SHINee"], "orange", "dotted", "SHINee")
plot_subplots(ax, fourth["Month"], fourth["EXO"], "blue", "dotted", "EXO")
plot_subplots(ax, fourth["Month"], fourth["NCT"], "green", "dashdot", "NCT")
plot_subplots(ax, fourth["Month"], fourth["SuperM"], "brown", "solid", "SuperM")
plot_subplots(ax, fourth["Month"], fourth["WAYV"], "purple", "dashdot", "WAYV")

plt.xticks(fourth["Month"])
ax.set_xticklabels(timenames, rotation = 90)

#annotating the events:
ax.annotate("EXO sub-unit", xy = (pd.Timestamp("2019-06-20"), 0.62), xytext = (pd.Timestamp("2019-01-01"), 0.47), arrowprops = {"arrowstyle":"->", "color":"red"})
ax.annotate("NCT EP", xy = (pd.Timestamp("2019-06-20"), 0.5), xytext = (pd.Timestamp("2019-01-01"), 0.35), arrowprops = {"arrowstyle":"->", "color":"red"})
ax.annotate("EXO cb.", xy = (pd.Timestamp("2019-11-01"), 0.7), xytext = (pd.Timestamp("2019-10-20"), 0.9), arrowprops = {"arrowstyle":"->", "color":"red"})

plt.legend(loc = "upper left")
plt.xlabel("Time (in Months)")
plt.ylabel("International Album Sales")


plt.show()
````
